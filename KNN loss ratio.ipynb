```python
import math
def FindLogPredictions(loss_ammount,loss_premium):
    log_predictions = []
    for ls,lp in zip(loss_ammount,loss_premium):
        
            log_predictions.append(math.log(ls) - math.log(lp))
       
    return log_predictions
```

```python
log_predictions = FindLogPredictions(final_df['SUM_Loss_ratio'],final_df['SUM_Annual_Premium'])
log_predictions
```

```python
# compute classification accuracy for the logistic regression model
from sklearn import metrics
# try K=1 through K=25 and record testing accuracy
k_range = range(1, 26)

# We can create Python dictionary using [] or dict()
scores = []

# We use a loop through the range 1 to 26
# We append the scores in the dictionary
for m in k_range:
    knn = neighbors.KNeighborsRegressor(m,weights="uniform")
    knn.fit(X_train, Y_train)
    Y_pred = knn.predict(X_test)
    scores.append(metrics.mean_absolute_error(Y_test, Y_pred))
    #print(metrics.mean_absolute_error(Y_test, Y_pred))

#print(scores)
```

```python
# import Matplotlib (scientific plotting library)
import matplotlib.pyplot as plt

# allow plots to appear within the notebook
%matplotlib inline

# plot the relationship between K and testing accuracy
# plt.plot(x_axis, y_axis)
plt.plot(k_range, scores)
plt.xlabel('Value of K for KNN')
plt.ylabel('Testing Error (Mean Absolute Error)')
```
